```python
# 传统写法
squares = []
for x in range(5):
    squares.append(x ** 2)
print(squares)  # [0, 1, 4, 9, 16]

# 列表推导式写法
squares = [x ** 2 for x in range(5)]
print(squares)  # [0, 1, 4, 9, 16]

# 带条件过滤：只保留偶数平方
even_squares = [x ** 2 for x in range(10) if x % 2 == 0]
print(even_squares)  # [0, 4, 16, 36, 64]

# 结合 if-else
results = ["偶数" if x % 2 == 0 else "奇数" for x in range(5)]
print(results)  # ['偶数', '奇数', '偶数', '奇数', '偶数']
```

```python
# -*- coding: utf-8 -*-
"""
day01_basics.py
包含三个基础练习题：
1. 斐波那契数列
2. 冒泡排序
3. 水仙花数判断
"""

def fibonacci(n):
    """
    生成斐波那契数列的前 n 项
    :param n: 正整数，数列长度
    :return: 包含前 n 项斐波那契数的列表
    """
    if n <= 0:
        return []
    elif n == 1:
        return [0]
    elif n == 2:
        return [0, 1]
    fib_list = [0, 1]
    for i in range(2, n):
        fib_list.append(fib_list[-1] + fib_list[-2])
    return fib_list


def bubble_sort(arr):
    """
    冒泡排序（原地修改）
    :param arr: 可变的列表
    :return: 排序后的列表（与原列表相同对象）
    """
    n = len(arr)
    for i in range(n - 1):
        # 标志位，用于优化：若本轮无交换则已有序
        swapped = False
        for j in range(n - 1 - i):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                swapped = True
        if not swapped:
            break
    return arr


def is_narcissistic(num):
    """
    判断一个整数是否为水仙花数（仅考虑三位数）
    水仙花数定义：一个 n 位数，其各位数字的 n 次幂之和等于该数本身。
    本例限定为三位数，即 100~999。
    :param num: 待判断的整数
    :return: 是水仙花数返回 True，否则返回 False
    """
    if not (100 <= num <= 999):
        return False
    # 提取百位、十位、个位
    hundreds = num // 100
    tens = (num // 10) % 10
    units = num % 10
    return hundreds ** 3 + tens ** 3 + units ** 3 == num


def find_all_narcissistic():
    """
    找出所有三位数中的水仙花数（100~999）
    :return: 水仙花数列表
    """
    return [num for num in range(100, 1000) if is_narcissistic(num)]


if __name__ == '__main__':
    print("=" * 40)
    print("1. 斐波那契数列（前10项）")
    fib_10 = fibonacci(10)
    print(f"斐波那契数列前10项: {fib_10}\n")

    print("=" * 40)
    print("2. 冒泡排序示例")
    test_list = [64, 34, 25, 12, 22, 11, 90]
    print(f"排序前: {test_list}")
    sorted_list = bubble_sort(test_list)
    print(f"排序后: {sorted_list}\n")

    print("=" * 40)
    print("3. 水仙花数判断")
    test_num=int(input("请输入一个三位数: "))
    print(f"{test_num} 是否水仙花数？{is_narcissistic(test_num)}")
    all_narc = find_all_narcissistic()
    print(f"所有三位水仙花数: {all_narc}")
    print("=" * 40)
```
